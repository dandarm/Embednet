```python
import json
with open('configs.json', 'r') as fp:
    configs = json.load(fp=fp)
```

```python
import matplotlib.pyplot as plt
losses = [c['test_loss'] for c in configs]
for l in losses:
    plt.plot(l)
plt.ylim(0, 0.001)
plt.show()
```

```python
import numpy as np
corrVSepoch = np.array([(c['training']['epochs'], np.mean(c['correlations'])) for c in configs])
plt.scatter(corrVSepoch.T[0], corrVSepoch.T[1]);
```
