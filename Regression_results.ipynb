# Test losses

```python
from scipy.optimize import curve_fit

def reverse_sigmoid(x, L, x0, k, b):
    y = L / (1 + np.exp(k*(x-x0))) + b
    return (y)


soglia_buona = 0.00002  # soglia sotto la quale la loss Ã¨ bassa e il training sembra essere andato bene
```

```python
mask_neurons = neurons_per_layer.apply(lambda x : len(x) == 3)
mask_nodes = df_regression_data['Num_nodes'] == 260
mask_lr = df_regression_data['learning_rate'] == 0.01
df_sel = df_regression_data[mask_neurons & mask_lr]
df_sel
```

```python
plt.figure(figsize=(20, 3))
for l in df_sel['test_loss']:
    plt.plot(l)
plt.ylim(0, 0.001)
plt.show()
```

```python
# fittiamo le learning curves

def fit_and_validate(data, debug=False):
    boolmask = np.isclose(data, np.median(data), rtol=1e-05, atol=1e-05)   
    try:
        mezzaltezza = int(np.median(np.where(boolmask)))
    except:
        plt.plot(data)
        mezzaltezza = np.median(range(len(data)))
    p0 = [max(data), mezzaltezza, 1, min(data)]
    
    success = False
    try:
        popt, pcov = curve_fit(reverse_sigmoid, list(range(len(data))), data, p0, method='dogbox')
        if debug: print(popt, pcov)
        success = True
    except:
        if debug: print("No fit convergence")
        return False

    if success:
        L, x0, k, b = popt
        if x0 < 0 or b > soglia_buona:  # aggiungere un check sulla differenza tra + e - infinito del fit risultatnte
            if debug: 
                print('Male?')
                return popt, pcov
            return False
        else:
            if debug: return popt, pcov
            return True
        
#for i in range(len(df_sel)):
df_regression_data['training_is_good'] = df_regression_data['test_loss'].apply(fit_and_validate)
```

```python
goods = df_regression_data[df_regression_data['training_is_good'] == True]
bads = df_regression_data[df_regression_data['training_is_good'] == False]

plt.figure(figsize=(20, 3))
print(f'{len(goods)} good training')
for l in goods['test_loss']:
    plt.plot(l)
plt.ylim(0, 0.001)
plt.show()

print(f'{len(bads)} bad training')
plt.figure(figsize=(20, 3))
for l in bads['test_loss']:
    plt.plot(l)
plt.ylim(0, 0.001)
plt.show()
```

```python
ii = [10, 28, 41]
for i in ii: # range(len(goods)):

    #print(i)
    data = goods['test_loss'].iloc[i]
    popt, pcov = fit_and_validate(data, debug=True)

    x = np.linspace(0, len(data), len(data))
    ys = np.array([reverse_sigmoid(xx, *popt) for xx in x])

    plt.plot(data)
    plt.plot(ys)
    #plt.plot(data-ys, '.')
    #plt.ylim(0,0.006)
    plt.show()
```
