# Vedo con tanti nodi e tanti grafi per tipo, addestrando tante epoche se la correlazione converge a un valore costante

```python
config_file = "configurations/classification_cm.yml"
config_class = Config(config_file)
trainer = Trainer(config_class)
```

```python
# inizializzo dei pesi che riuserò
train_with_same_dataset = False
train_with_same_initial_weightparameter = False
if train_with_same_dataset:
    trainer.init_dataset(parallel=True)
    trainer.load_dataset(trainer.gg.dataset)
if train_with_same_initial_weightparameter:
    trainer = init_all_config_class(config_class)
    saved_initial_weights = new_parameters(trainer.model, method=Inits.xavier_uniform)
```

```python
diz_trials = {'graph_dataset.Num_grafi_per_tipo': [300]*3}
gc = GridConfigurations(config_class, diz_trials)
gc.make_configs()
len(gc.configs)
```

```python
for i, c in enumerate(gc.configs):
    print(f'Run {i+1}/{len(gc.configs)}')
    try:         
        trainer.reinit_conf(c)
        
        if not train_with_same_dataset:
            trainer.init_dataset()
            trainer.load_dataset(trainer.gg.dataset)
        if train_with_same_initial_weightparameter:
            model = trainer.init_GCN(saved_initial_weights)
            trainer.load_model(model)
        else:
            model = trainer.init_GCN()
            trainer.load_model(model)
         
        trainer.launch_training()        
        # serve per get_corrs_training
        experiments.t = trainer
        experiments.c = c
        avg_corr_classes = get_corrs_training()
        avg_corr_classes = np.array(avg_corr_classes).T
    
        gc.config_dataframe.at[i, ('risultati','test_loss')] = trainer.test_loss_list
        gc.config_dataframe.at[i, ('risultati','test_accuracy')] = trainer.accuracy_list
        gc.config_dataframe.at[i, ('risultati','correlation_allclasses')] = avg_corr_classes.tolist()
        
        gc.config_dataframe.loc[i, ('risultati','data')] = pd.Timestamp.now()
        
        #gc.config_dataframe.iloc[i:i+1].to_csv(outfile, index=False, mode='a', header=not os.path.exists(outfile))

    except Exception as e: 
        print(repr(e))
        raise(e)
```
